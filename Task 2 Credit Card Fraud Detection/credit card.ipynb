```python
X=df.drop(['is_fraud'],axis=1)
Y=df['is_fraud']
```

```python
from sklearn.preprocessing import OrdinalEncoder
cols =['trans_date_trans_time','merchant','category','first','last','gender','street','city','state','job','dob','trans_num']
encoder =OrdinalEncoder()
X[cols]=encoder.fit_transform(X[cols])
```

```python
Y= df[['is_fraud']].values
```

```python
df['is_fraud'].value_counts()
```

```python
print('Independent fet',X.shape)
print('Dependent fet',Y.shape)
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
smt=SMOTE()
```

```python
x,y =smt.fit_resample(X,Y)
```

```python
print('Data  :',x.shape)
print('Data  :',y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size =0.2,random_state=42)
```

```python
LR=LogisticRegression()
```

```python
LR.fit(x_train,y_train)
```

```python
pred_Log =LR.predict(x_test)
```

```python
pred_test=LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('LR Accuracy: ',accuracy_score(y_test,pred_Log))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(max_depth =50,random_state =42)
```

```python
dtc.fit(x_train,y_train)
```

```python
pred_dtc =dtc.predict(x_test)
print('DTC Accuracy :',accuracy_score(y_test,pred_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=20, random_state=2)
rfc.fit(x_train, y_train)
```

```python
pred_rfc =rfc.predict(x_test)
print('RFC accuracy',accuracy_score(y_test,pred_rfc))
```
